```python
import pandas as pd
```

```python
data_path = 'data/origin/'
df1 = pd.read_csv(data_path + '01_Monday-WorkingHours.pcap_ISCX.csv')
df2 = pd.read_csv(data_path + '02_Tuesday-WorkingHours.pcap_ISCX.csv')
df3 = pd.read_csv(data_path + '03_Wednesday-workingHours.pcap_ISCX.csv')
df4 = pd.read_csv(data_path + '04_Thursday-WorkingHours-Afternoon-Infilteration.pcap_ISCX.csv')
df5 = pd.read_csv(data_path + '04_Thursday-WorkingHours-Morning-WebAttacks.pcap_ISCX.csv', low_memory=False)
df6 = pd.read_csv(data_path + '05_Friday-WorkingHours-Afternoon-DDos.pcap_ISCX.csv')
df7 = pd.read_csv(data_path + '05_Friday-WorkingHours-Afternoon-PortScan.pcap_ISCX.csv')
df8 = pd.read_csv(data_path + '05_Friday-WorkingHours-Morning.pcap_ISCX.csv')

df = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8])
df.info()
```

```python
df.tail()
```

```python
df.isnull().sum()
df.dropna()

df["Label"] =  df[" Label"]
df.drop(columns=' Label', inplace=True)
```

```python

label_counts = df["Label"].value_counts()
label_counts
```

```python
df.columns = df.columns.str.strip()
df["Label"] = df["Label"].str.replace(' � ','-',regex=True)
df["Label"].unique()
```

```python
min_data_counts = 5000
abnormal_label_counts = label_counts.loc[label_counts > min_data_counts] - 1
print("Abnormal label count: %d"%(len(label_counts) - 1))
print("Abnormal label count (%d+ rows): %d"%(min_data_counts, len(abnormal_label_counts)))
```

```python
normal_sample_count = min_data_counts * 2
print("Normal data count:", normal_sample_count)
print("Abnormal data count:", min_data_counts*len(abnormal_label_counts))
print("Total count:", normal_sample_count + min_data_counts*len(abnormal_label_counts))
```

```python
def get_data(label, size):
    df_label = df[df['Label'] == label]
    return df_label.sample(n=min(len(df), size))

abnormal_samples = []
for label in abnormal_label_counts.index:
        abnormal_samples.append(get_data(label, min_data_counts))

normal_sample = get_data('BENIGN', normal_sample_count)
abnormal_sample = pd.concat(abnormal_samples)
```

```python
df_compact = pd.concat([normal_sample, abnormal_sample], ignore_index=True)
df_compact.info()
```
